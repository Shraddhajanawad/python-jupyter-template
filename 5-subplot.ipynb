### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 10)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Chart
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Chart")

# Bar Chart
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title("Bar Chart")

# Scatter Plot
axs[1, 0].scatter(x, y, color='r', marker='x')
axs[1, 0].set_title("Scatter Plot")

# Pie Chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'lightcoral'])
axs[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```
